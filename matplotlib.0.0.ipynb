##### **Matplotlib v.1**

```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib as mpl 
rcParamsInicial = dict(mpl.rcParams)
```

```python
import pandas as pd

def checa_params(dicionario1, nome1, dicionario2, nome2):
  df = pd.DataFrame(columns=["chave", nome1, nome2])
  for chave in dicionario1.keys():
    if dicionario1[chave] != dicionario2[chave]:
      df = df.append(pd.DataFrame({"chave":[chave], 
                                   nome1:[dicionario1[chave]], 
                                   nome2:[dicionario2[chave]]}), ignore_index=True)
  return(df)

checa_params(rcParamsInicial, "rcParamsInicial", mpl.rcParamsDefault, "rcParamsDefault")
```

```python
x = np.linspace(0,2*np.pi,1000)
y = np.sin(x)
z = np.cos(x)
```

```python
fig,ax = plt.subplots(nrows=2, ncols=1)
fig.suptitle("Título do Conjunto")
ax[0].scatter(x, y, color='red')
ax[0].set_title("Gráfico 1")
ax[1].scatter(x, z, color='blue')
ax[1].set_title("Gráfico 2")
plt.show()
```

```python
plt.style.use('default')
fig,ax = plt.subplots(nrows=2, ncols=1, dpi=72, figsize=(6,4))
fig.suptitle("Título do Conjunto")
ax[0].scatter(x, y, color='red')
ax[0].set_title("Gráfico 1")
ax[1].scatter(x, z, color='blue')
ax[1].set_title("Gráfico 2")
plt.show()
```

```python
mpl.rcParams.update(rcParamsInicial)
```

##### **Matplotlib v.2**

```python
import numpy as np
from matplotlib import pyplot as plt
plt.style.use("ggplot")
import matplotlib as mpl 
rcParamsInicial = dict(mpl.rcParams) # figure.dpi=72, figure.figsize=(6,4)
```

```python
x = np.linspace(0,2*np.pi,1000)
y = np.sin(x)
z = np.cos(x)
```

```python
fig,ax = plt.subplots(nrows=2, ncols=1)
fig.suptitle("Título do Conjunto", y=1.04)
ax[0].scatter(x, y, color='red')
ax[0].set_title("Gráfico 1")
ax[1].scatter(x, z, color='blue')
ax[1].set_title("Gráfico 2")
fig.tight_layout()
```
