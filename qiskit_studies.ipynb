```python
from qiskit.providers.aer import AerSimulator
sim = AerSimulator() 
```

```python
from qiskit import QuantumCircuit
qc = QuantumCircuit(4,4)
qc.x([0,1])
qc.cx([0,1],[2,2])
qc.ccx(0,1,3)
qc.measure([0,1,2,3],[0,1,2,3])
display(qc.draw(output='mpl'))

job = sim.run(qc)
result = job.result()
result.get_counts()
```

```python
from qiskit import QuantumCircuit
qc = QuantumCircuit(1,1)
qc.x(0)
qc.h(0)
qc.h(0)
qc.measure(0,0)
display(qc.draw(output='mpl'))

job = sim.run(qc)
result = job.result()
result.get_counts()
```
